```python
from pathlib import Path
```

```python
single_image_path = Path('../../tests/data/mixed_images/ukbench00120.jpg')
image_dir = Path('../../tests/data/mixed_images')
```

# Hashing

```python
from imagededup.hashing import Hashing
```

### Hash of a single image

```python
myhasher = Hashing(method='phash')
hash_string = myhasher.hash_image(single_image_path) # get_hash
```

```python
hash_string
```

##### average hash

```python
myhasher = Hashing(method='ahash')
hash_string = myhasher.hash_image(single_image_path)
```

```python
hash_string
```

## Hash on a directory of images 

```python
myhasher = Hashing(method='phash')
hash_dict = myhasher.hash_dir(image_dir)
```

```python
hash_dict
```
